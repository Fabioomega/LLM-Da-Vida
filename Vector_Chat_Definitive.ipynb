# QA Def

```python
from typing import TypedDict
from collections import deque


class QA(TypedDict):
    """Uma QA é um conjunto de uma pergunta e uma resposta que estão associadas"""

    question: str
    answer: str


def create_QA(question: str, answer: str):
    """Cria uma QA como se fosse uma classe e não um dicionario para ficar mais facil de usar"""

    return QA(question=question, answer=answer)


class QAList:
    """
    Uma QAList é uma lista de QA que pode ou não ter um limite de perguntas e respotas.
    Para mostrar todas as conversas basta chamar conversation_history.
    A lista é organizada em forma de chegada, sendo o útlimo adiciondo o útlimo a aparecer quando a conversa inteira é mostrada.
    Caso um novo elemento seja adicionado e o limite tenha sido atingido o elemento mais antigo é removido.
    """

    def __init__(self, limit: int = -1) -> None:
        self.qalist: deque[QA] = deque([])
        if limit == 0:
            raise ValueError("The limit must be different than zero!!")
        self.limit = limit

    def conversation_history(self):
        """Retorna a conversa inteira contando todas as perguntas e respostas"""

        f = ""

        for qa in self.qalist:
            f += f"{qa['question']}\n{qa['answer']}\n\n"

        f = f[:-2]  # Remove os dois ultimos \n

        return f

    def conversation_history_slice(self, beg: int = 0, end: int = None):
        """Retorna um trecho da conversa. Indexe como uma lista"""

        f = ""

        if end is None or end > self.__len__():
            end = self.__len__()

        for i in range(beg, end):
            f += f"{self.__getitem__(i)['question']}\n{self.__getitem__(i)['answer']}\n\n"

        f = f[:-2]  # Remove os dois ultimos \n

        return f

    def add(self, new_qa: QA):
        """Adiciona um novo elemento na lista"""

        if self.limit < 0 or len(self.qalist) < self.limit:
            self.qalist.append(new_qa)
        else:
            self.qalist.popleft()
            self.qalist.append(new_qa)

    def clear(self):
        """Limpa a lista"""
        self.qalist.clear()

    def __len__(self):
        return len(self.qalist)

    def __getitem__(self, subscript):
        if isinstance(subscript, slice):
            subscript: slice
            raise NotImplementedError(
                "Ops... Slice was not implemented!!! Maybe use conversation_history_slice instead!"
            )
        else:
            return self.qalist[subscript]
```

# Fake Pipeline Because LangChain Hates Me

```python
class FakePipeline:

  def __init__(self, pipe, generation_kwargs):
    self.task = 'text-generation'
    self.pipe = pipe
    self.generation_kwargs = generation_kwargs

  def __call__(self, *args, **kwargs):
    return self.pipe(*args, **kwargs, **self.generation_kwargs)

def create_special_pipeline(pipe, generation_kwargs):
    return FakePipeline(pipe, generation_kwargs)
```

# Adding to the store

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import TextLoader


def add_txt_to_db(filename: str, chunk_size: int, chunk_overlap: int) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    loader = TextLoader(f"{filename}.txt")
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return db.add_documents(text_splitter.split_documents(documents))

def add_texts_to_db(texts: list[str]) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    return db.add_texts(texts)

def clear_db():
    db.delete_collection()
```

# Retrieval Chain Output

```python
from typing import Optional

class RetrievalResult(TypedDict):
    debug: Optional[str]
    result: str

class CompleteGenerationResult(TypedDict):
    conversation_history: QAList
    result: RetrievalResult
```

# Create Chat Retrieval Chain

```python
from langchain import PromptTemplate


TEMPLATE = """Given the following conversation and a follow up question, rephrase the follow up question in portuguese to be a standalone question.
Remember that Universidade Federal de Santa Catarina is commonly called UFSC.

Chat History:
{chat_history}
Follow Up Input: {question}
Standalone Question in Portuguese:"""

PT_TEMPLATE = """Dada a conversa a seguir e uma pergunta complementar, reformule a pergunta complementar para ser uma pergunta independente.
Lembre-se que Universidade Federal de Santa Catarina é chamada de UFSC.

Conversa:
{chat_history}
Pergunta Complementar: {question}
Pergunta Independente:"""

standalone_question_template = PromptTemplate.from_template(PT_TEMPLATE)



def create_standalone_question(query: str, ch: QAList, generation_kwargs: dict):
    if len(ch) == 0:
        return query

    formatted_text = standalone_question_template.format(question=query, chat_history=ch.conversation_history())

    output = generate_text(formatted_text, **generation_kwargs)[0]['generated_text']

    return str(output[len(formatted_text):])

def generate_text_with_RetrievalQA(query: str, generation_kwargs: dict) -> RetrievalResult:
    docs = db.similarity_search(query)

    if len(docs) == 0:
      return {'result': 'Desculpa, não sei hehe...'}

    return {'result': qa_chain.run(query)}

def generate_text_with_RetrievalQA_DEBUG(query: str, generation_kwargs: dict) -> RetrievalResult:
    docs = db.similarity_search(query)

    if len(docs) == 0:
      return {'debug': 'Não houveram queries hehe...\n\nResposta: Desculpa, não sei hehe...',
              'result': 'Desculpa, não sei hehe...'}

    answer = qa_chain.run(query)

    formatted_docs = ''

    for i, e in enumerate(docs):
        formatted_docs += f'{i + 1}. ' + e.page_content + '\n'

    return {'debug' : f"A query mais provavel foi:\n{formatted_docs}\n\nResposta: {answer}",
             'result': answer}

def generate_text_with_context(query: str, conversation_history: QAList, standalone_generation_kwargs: dict, retrieval_generation_kwargs: dict, debug=False) -> CompleteGenerationResult:
    standalone_answer = create_standalone_question(query, conversation_history, standalone_generation_kwargs)

    if debug:
      print('Standalone question:', standalone_answer)

    result = {}

    if debug:
        result = generate_text_with_RetrievalQA_DEBUG(standalone_answer, retrieval_generation_kwargs)
    else:
        result = generate_text_with_RetrievalQA(standalone_answer, retrieval_generation_kwargs)

    conversation_history.add(create_QA(query, result['result']))
    return {'result': result, 'conversation_history': conversation_history}
```

## Generation

```python
def generate(query: str, conversation: QAList, standalone_generation_kwargs: dict, retrieval_generation_kwargs: dict, debug=False) -> tuple[QAList, str]:
    response: CompleteGenerationResult = generate_text_with_context(query, conversation, standalone_generation_kwargs, retrieval_generation_kwargs, debug=debug)
    conversation = response['conversation_history']

    true_result: RetrievalResult = response['result']

    if debug:
        return (conversation, true_result['debug'])

    return (conversation, true_result['result'])

def testing_input(st: list[str], generation_args: dict, limit=3):
    conversation = QAList(limit)

    global qa_chain
    qa_chain = create_retrieval_chain(generation_args, score_threshold = 5.0)

    for query in st:
      conversation, output = generate(query, conversation, generation_args, generation_args, True)
      print(output)
```
