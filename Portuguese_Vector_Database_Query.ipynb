```python
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.document_loaders import TextLoader
from langchain.document_loaders import PyPDFLoader

FILENAME = 'UFSC pt.txt'

# load the document and split it into chunks
loader = TextLoader(f"/content/{FILENAME}")
documents = loader.load()

#loader = PyPDFLoader(f"/content/{FILENAME}")
#documents = loader.load()
documents = loader.load_and_split()

# split it into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=64)
docs = text_splitter.split_documents(documents)

# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="rufimelo/bert-large-portuguese-cased-sts")

# load it into Chroma
db = Chroma.from_documents(docs, embedding_function)
```

Inicializa a pipeline do modelo do Bloom 3b

```python
import torch
from transformers import pipeline

generate_text = pipeline("text-generation", model="bigscience/bloom-3b", torch_dtype=torch.bfloat16,
                         trust_remote_code=True, device_map="auto", return_full_text=True, max_length=600)
```

Cria um template com Langchain e usa a pipeline e o banco vetorial para criar uma cadeia de RetrievalQA.

Por padrão o k da chain é 4. Que coincide com o padrão de query do Chroma

A LLM não está usando chat. Para esse modelo e processamento do Collab qualquer modelo de chat é bem lento

```python
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

template_pt = """Use as proximas partes como contexto para responder a pergunta no final.
Se você não souber a resposta, apenas diga que você não sabe, não tente inventar uma resposta.

{context}

Pergunta: {question}
Resposta Útil:"""

template = """Use the following pieces of context to answer the question at the end.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

{context}

Question: {question}
Helpful Answer:"""

QA_CHAIN_PROMPT = PromptTemplate.from_template(template)

hf_pipeline = HuggingFacePipeline(pipeline=generate_text)

qa_chain = RetrievalQA.from_chain_type(
hf_pipeline,
retriever=db.as_retriever(),
chain_type_kwargs={"prompt": QA_CHAIN_PROMPT})
```

Inicializa uma função para mostrar os dados utilizados pelo modelo para criar a resposta seguindo o modelo da template e a resposta do modelo as instruções.

```python
def ask_and_query(text: str) -> str:
  answer = qa_chain({"query": text})

  docs = db.similarity_search(text)

  formatted_docs = ''

  for i, e in enumerate(docs):
    formatted_docs += f'{i + 1}. ' + e.page_content + '\n'

  return f"The probable query result was:\n{formatted_docs}\n\nAnswer: {answer['result']}"

def query(text: str) -> str:
  docs = db.similarity_search(text)

  formatted_docs = ''

  for i, e in enumerate(docs):
    formatted_docs += f'{i + 1}. ' + e.page_content + '\n'

  return formatted_docs
```

```python
print(query('O que você sabe sobre a Universidade Federal de Santa Catarina?'))
```

```python
print(ask_and_query('O que significa UFSC?'))
```
