# Translation

```python
import translators as ts

SERVICE = "google"


def translate_to_pt(text: str) -> str:
    return ts.translate_text(
        text, from_language="en", to_language="pt", translator=SERVICE
    )


def translate_to_en(text: str) -> str:
    return ts.translate_text(
        text, from_language="pt", to_language="en", translator=SERVICE
    )
```

# LLM Def

```python
from langchain.llms import CTransformers
config = {'max_new_tokens': 256, 'temperature': 0.8, 'repetition_penalty': 1.1, 'context_length': 2024, 'gpu_layers': 10_000_000_000, 'batch_size': 8}

llm = CTransformers(model='TheBloke/Llama-2-7b-Chat-GGUF', model_file='llama-2-7b-chat.Q6_K.gguf', model_type='llama', config=config)
```

# Embeddings Def

```python
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
embedding_function = SentenceTransformerEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

# Vectorstore Def

```python
from langchain.vectorstores import Chroma

db = Chroma('llm_store', embedding_function)
```

# Adding to the store

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import TextLoader


def add_txt_to_db(filename: str, chunk_size: int, chunk_overlap: int) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    loader = TextLoader(f"{filename}.txt")
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return db.add_documents(text_splitter.split_documents(documents))

def add_pt_txt_to_db(filename: str, chunk_size: int, chunk_overlap: int) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    loader = TextLoader(f"{filename}.txt")
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    documents = []
    for doc in text_splitter.split_documents(documents):
      doc.page_content = translate_to_en(doc.page_content)
      documents.append(doc)

    return db.add_documents(documents)

def add_texts_to_db(texts: list[str]) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    return db.add_texts(texts)

def add_pt_texts_to_db(text: list[str]) -> list[str]:
    """Retorna os ids dos documentos adicionados"""

    translated = []
    for el in text:
      translated.append(translate_to_en(el))
    return add_texts_to_db(translated)

def clear_db():
    db.delete_collection()
```

# Create RetrievalQA Chain

```python
from langchain import PromptTemplate
from langchain.chains import ConversationalRetrievalChain

test_template = """[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
Please ensure that your responses are socially unbiased and positive in nature. If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct.
If you don't know the answer to a question, please don't share false information.
If the question is not about the Federal University of Santa Catarina, politely inform them that you are tuned to only answer questions about Federal University of Santa Catarina.
You will always call Federal University of Santa Catarina as UFSC.
<</SYS>>
Use the following pieces of context to answer the question at the end.
{context}
Question: {question}
[/INST]
"""

standalone_test_question_template = """[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
Please ensure that your responses are socially unbiased and positive in nature.
You will always call Federal University of Santa Catarina as UFSC.
<</SYS>>
Summarize the following conversation and a follow up question and rephrase it to be a standalone question.
Chat History:
{chat_history}
Follow Up Input: {question}
[/INST]
"""


def create_retrieval_chain(score_threshold: float = 0.0) -> ConversationalRetrievalChain:

    QA_CHAIN_PROMPT = PromptTemplate.from_template(test_template)

    CONDENSE_PROMPT = PromptTemplate.from_template(standalone_test_question_template)

    return ConversationalRetrievalChain.from_llm(
    llm,
    retriever=db.as_retriever(search_kwargs={'score_threshold': score_threshold}),
    condense_question_prompt=CONDENSE_PROMPT,
    combine_docs_chain_kwargs={"prompt": QA_CHAIN_PROMPT},
    verbose=True
)

qa_chain = create_retrieval_chain(score_threshold = 1.2)
```

# Testing and Generating Text

## Generation

```python
def generate(query: str, chat_history) -> str:
    return qa_chain({'question': query, 'chat_history': chat_history})

def testing_input(st: list[str], generation_args: dict):
    change_generation_kwargs(generation_args)

    chat_history = []

    for query in st:
      print('Q:', query)
      output = generate(query, chat_history)
      print('A:', output['answer'])
      chat_history.append((query, output['answer']))
```
